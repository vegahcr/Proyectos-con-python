```python
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
X1=np.array([2.1,1.6,1.4,1.3,1.25,1.2,1.01,0.8])
Y1=np.array([10, 20, 30, 40, 50,  60, 100, 200])
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(X1, Y1, color='b')
ax.set_xlabel('Rs/Ro')
ax.set_ylabel('PPM')
ax.set_title('MQ-135. Curva de sensibilidad')
plt.show()
```

```python
X = X1[:,np.newaxis]
Y = Y1[:,np.newaxis]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X,Y)
```

```python
regr.coef_
```

```python
m = regr.coef_[0]
b = regr.intercept_
b
```

```python
Y_pred = regr.predict(X)
```

```python
y = m*X+b
```

Ecuación de regresión lineal,   y = -127.139x + 233.1628

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(X1, Y1, color='b', marker='x', label='Datos del fabricante')
ax.scatter(X1, Y_pred, color='r',label='Predicciones')
ax.set_xlabel('Rs/Ro')
ax.set_ylabel('PPM')
ax.set_title('Regresión lineal para el sensor MQ-135')
ax.plot(X1,y,color = 'r')
plt.legend()
plt.show()
```

```python
r2_score(Y1, Y_pred)
```
