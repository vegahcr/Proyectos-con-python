```python
df.columns = ['#', 'ID Cliente', 'Genero', 'Jubilado', 'En matrimonio',
       'Dependiente', 'Antiguedad(meses)', 'Servicio telefónico', 'Multiples lineas',
       'Servicio de Internet', 'Navegación segura', 'Copia de seguridad', 'Proteción de datos',
       'Soporte técnico', 'Servicio de Streaming', 'Cine por Streaming', 'Tipo de contrato',
       'Facturación electrónica', 'Método de pago', 'Pago mensual', 'Totalhistórico',
       'Abandono']
```

```python
df.dtypes
```

Eliminar filas con datos nulos

```python
df.drop(df[df['Totalhistórico']==" "].index,inplace = True)
```

Eliminar columnas con datos inecesarios

```python
df.drop(['#','ID Cliente'],axis=1,inplace=True)
```

```python
df.head(4)
```

Convertir las columnas con datos que contienen palabras unicas a datos numericos

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
columnas = ['Genero', 'Jubilado', 'En matrimonio',
       'Dependiente', 'Servicio telefónico', 'Multiples lineas',
       'Servicio de Internet', 'Navegación segura', 'Copia de seguridad', 'Proteción de datos',
       'Soporte técnico', 'Servicio de Streaming', 'Cine por Streaming', 'Tipo de contrato',
       'Facturación electrónica','Método de pago','Abandono']
```

```python
for i in columnas:
    df[i] = LabelEncoder().fit_transform(df[i])
```

```python
df.head(4)
```

```python
df.dtypes
```

```python
df["Totalhistórico"] = pd.to_numeric(df["Totalhistórico"], downcast="float")
```

```python
df.dtypes
```

```python
df
```

Entrenar el modelo de regresión logistica

```python
y = df['Abandono'] #Columna objetivo
```

```python
X = df.drop('Abandono', axis=1) #Caracteristicas
```

```python
from sklearn.model_selection import train_test_split
#70% entrenamiento, 30% prueba, toma los datos de forma aleatoria
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelo = LogisticRegression()
modelo.fit(X_train, y_train)
```

```python
predicciones = modelo.predict(X_test)
```

```python
predicciones #Valores con prediccion
```

```python
y_test #Valores reales
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predicciones))
```

precision de 88% para clientes actuales
